```python
import json
import os

DATA_FILE = 'users.json'

# Load user data
def load_users():
    if os.path.exists(DATA_FILE):
        with open(DATA_FILE, 'r') as f:
            return json.load(f)
    else:
        return {}

# Save user data
def save_users(users):
    with open(DATA_FILE, 'w') as f:
        json.dump(users, f, indent=4)

# Header
def print_header():
    print("="*40)
    print("       ðŸ¦ Welcome to Python ATM       ")
    print("="*40)

# Ask if user has account
def ask_account():
    while True:
        response = input("Do you have an account? (yes/no): ").strip().lower()
        if response in ['yes', 'no']:
            return response
        else:
            print("â— Please answer with 'yes' or 'no'.")

# Register a new user
def register(users):
    while True:
        new_pin = input("Set a new 4-digit PIN: ").strip()
        if not (new_pin.isdigit() and len(new_pin) == 4):
            print("âŒ PIN must be exactly 4 digits.")
            continue
        if new_pin in users:
            print("âš ï¸ This PIN already exists. Try a different one.")
            continue
        break
    
    while True:
        deposit = input("Enter initial deposit amount: ").strip()
        if deposit.isdigit() and int(deposit) > 0:
            deposit = int(deposit)
            break
        else:
            print("âŒ Invalid deposit amount. Must be a number > 0.")

    users[new_pin] = {"balance": deposit}
    save_users(users)
    print(f"âœ… Account created successfully! Your PIN is {new_pin}")
    return new_pin

# Authenticate user
def authenticate(users):
    attempts = 3
    while attempts > 0:
        pin = input("Enter your 4-digit PIN: ").strip()
        if pin in users:
            print("âœ… Login successful!\n")
            return pin
        else:
            attempts -= 1
            print(f"âŒ Incorrect PIN. {attempts} attempt(s) left.")
    print("ðŸ”’ Account locked.")
    return None

# ATM Menu
def atm_menu():
    print("\nSelect an option:")
    print("1. ðŸ’° Check Balance")
    print("2. âž• Deposit Money")
    print("3. âž– Withdraw Money")
    print("4. ðŸšª Exit")

# ATM Operations
def atm(pin, users):
    while True:
        atm_menu()
        choice = input("Enter choice (1-4): ")

        if choice == '1':
            print(f"ðŸ’µ Current Balance: Rs {users[pin]['balance']}")
        
        elif choice == '2':
            amount = input("Enter deposit amount: ")
            if amount.isdigit() and int(amount) > 0:
                users[pin]['balance'] += int(amount)
                save_users(users)
                print(f"âœ… Deposited Rs {amount}. New Balance: Rs {users[pin]['balance']}")
            else:
                print("âš ï¸ Invalid amount.")
        
        elif choice == '3':
            amount = input("Enter withdrawal amount: ")
            if amount.isdigit() and int(amount) > 0:
                amount = int(amount)
                if amount <= users[pin]['balance']:
                    users[pin]['balance'] -= amount
                    save_users(users)
                    print(f"âœ… Withdrawn Rs {amount}. Remaining Balance: Rs {users[pin]['balance']}")
                else:
                    print("âŒ Insufficient funds.")
            else:
                print("âš ï¸ Invalid amount.")
        
        elif choice == '4':
            print("ðŸ‘‹ Thank you for using Python ATM!")
            break
        
        else:
            print("âŒ Invalid option.")

# Main program
def main():
    print_header()
    users = load_users()
    
    response = ask_account()
    if response == 'no':
        pin = register(users)
    else:
        pin = authenticate(users)
    
    if pin:
        atm(pin, users)

if __name__ == "__main__":
    main()
```

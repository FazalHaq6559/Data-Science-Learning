## âœ… Personal Finance Management System


```python
print("==== Personal Finance Tracker ====")
balance = 0
transactions = []  # Each transaction is a dictionary with 'type' and 'amount'

while True:
    print("------------------------")
    print("\nPlease select an option:")
    print("1. Add income")
    print("2. Add expense")
    print("3. View transaction history")
    print("4. View balance")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == '1':
        amount = float(input("Enter income amount: "))
        balance += amount
        transactions.append({
            'type': 'Income',
            'amount': amount
        })
        print("------Income added------")

    elif choice == '2':
        amount = float(input("Enter expense amount: "))
        if amount > balance:
            print("------Insufficient balance------")
        else:
            balance -= amount
            transactions.append({
                'type': 'Expense',
                'amount': amount
            })
            print("------Expense added------")

    elif choice == '3':
        print("\nTransaction History:")
        if not transactions:
            print("------No transactions yet------")
        else:
            for t in transactions:
                print(f"[Transiction Type : {t['type']}\tTransiction Amount: {t['amount']}]")

    elif choice == '4':
        print(f"Current Balance: {balance}")

    elif choice == '5':
        print("------Exiting the program. Goodbye!------")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 5.")
```

### 3. Explain the difference between np.array(), np.zeros(), np.ones(), and np.empty().


**np.array()** Creates a NumPy array from an existing list or tuple.

 - **Use:** When you already have data in Python and want to work with it efficiently using NumPy operations.

**np.zeros()** → Creates a new array of a given shape, filled with zeros.

- **Use:** Useful for initializing empty data structures, placeholders, or weight matrices in ML.

**np.ones()** Creates a new array of a given shape, filled with ones.

- **Use:** Often used in mathematical operations, masks, or as a starting matrix for computations.

**np.empty()** Creates a new array of a given shape without initializing values (memory garbage).

- **Use:** When you only care about speed and will overwrite the array values immediately.

### 9. Normalize an array of exam scores (values between 0–100) into a range of 0–1 using NumPy.


```python
# for 0-1 we use min-max scaling
scores = np.array([45, 67, 89, 100, 56])
normalized = (scores - scores.min()) / (scores.max() - scores.min())
normalized
```

### 12. How do you calculate the correlation matrix of a dataset with NumPy?


```python
data = np.random.rand(100, 3)  # 100 samples, 3 features

corr_matrix = np.corrcoef(data, rowvar=False)

print("Correlation Matrix:\n", corr_matrix)
```

### 14. Simulate a dataset of 500 coin tosses using np.random.choice() and calculate the probability of heads.



```python
tosses = np.random.choice(['H', 'T'], size=500, p=[0.5, 0.5])
prob_heads = np.sum(tosses == 'H') / len(tosses)

print("Probability of Heads:", prob_heads)
```

### 15. Standardize a dataset (zero mean and unit variance) using NumPy.


```python
data = np.random.randint(1, 100, (10, 3))  # 10 rows, 3 features

standardized = (data - np.mean(data, axis=0)) / np.std(data, axis=0)

print("Standardized:\n", standardized)
```

### 17. Compute eigenvalues and eigenvectors of a covariance matrix. Explain their role in PCA (Principal Component Analysis).


**Eigenvalues** represent the amount of variance explained by each eigenvector (direction).

**Eigenvectors** directions in which the data varies the most (principal components).

```python
# Covariance matrix
X = np.random.rand(5, 3)
cov = np.cov(X.T)

eigvals, eigvecs = np.linalg.eig(cov)

print("Eigenvalues:", eigvals)
print("Eigenvectors:\n", eigvecs)
```

### 18. Solve a system of linear equations using np.linalg.solve().

**Example:** Solve system

2x + y = 5

x - y = 1


```python
A = np.array([[2, 1], [1, -1]])
b = np.array([5, 1])

solution = np.linalg.solve(A, b)
print("Solution (x, y):", solution)
```

### 19. Find the rank, determinant, and inverse of a matrix using NumPy.

**Explanation:**

**Rank** number of independent rows/columns

**Determinant** tells if matrix invertible

**Inverse** used in solving equations

```python
M = np.array([[4, 7], [2, 6]])

print("Rank:", np.linalg.matrix_rank(M))
print("Determinant:", np.linalg.det(M))
print("Inverse:\n", np.linalg.inv(M))
```

### 20. Load a grayscale image as a NumPy array and apply thresholding: set all pixel values above 128 to 255 and below to 0.

```python
import numpy as np
from PIL import Image

# Load image as grayscale
img = Image.open("download.webp").convert("L")
arr = np.array(img)

# Thresholding
thresholded = np.where(arr > 128, 255, 0)

Image.fromarray(np.uint8(thresholded)).save("thresholded.png")


import matplotlib.pyplot as plt
img = Image.open("thresholded.png")

plt.imshow(img, cmap="gray")
plt.axis("off") 
plt.show()
```

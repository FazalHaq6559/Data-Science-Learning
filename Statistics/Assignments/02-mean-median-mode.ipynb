<b style="color:blue"> Q1. What is the difference between population mean and sample mean? Provide a real-world example where each is used.</b>

| Aspect      | Population Mean | Sample Mean |
|:-------------|:-----------------|:-------------|
| **Definition**  | The average of all values in the entire population. | The average of values taken from a subset (sample) of the population. |
| **Formula**     | μ = (ΣX) / N    | x̄ = (ΣX) / n |
| **Purpose**     | Describe the true average of the population. | Estimate population mean when collecting all data is impractical. |

**Example:**  
- **Population mean:** Average age of all citizens in a country.  
- **Sample mean:** Average age of citizens from a survey of 1,000 people.

 <b style="color:blue"> Q4. Explain when the median is a better measure than the mean in data analysis.

Median is better when the dataset has **outliers** or is **skewed**.  
- Mean is sensitive to extreme values.  
- Median, as the middle value, is unaffected by extremes and gives a more accurate central tendency.

<b style="color:blue">Q6. Write your own logic (without using built-in functions) to calculate the mean of 10 numbers.

```python
numbers = [5, 8, 12, 7, 15, 10, 9, 6, 11, 14]
total = 0
count = 0

for num in numbers:
    total += num
    count += 1

mean = total / count
print("Mean:", mean)
```

<b style="color:blue">Q7. Write your own logic to calculate the median of a list of 7 numbers sorted in ascending order.

```python
numbers = [5, 8, 12, 15, 18, 21, 24]  # already sorted
n = len(numbers)

if n % 2 == 1:  # odd length
    median = numbers[n // 2]
else:  # even length
    median = (numbers[(n//2)-1] + numbers[n//2]) / 2

print("Median:", median)
```

<b style="color:blue"> Q8. Write your own logic to find the mode of a list of 10 numbers.

```python
numbers = [2, 4, 2, 6, 4, 4, 8, 2, 4, 6]
frequency = {}
max_count = 0
mode = None

for num in numbers:
    frequency[num] = frequency.get(num, 0) + 1
    if frequency[num] > max_count:
        max_count = frequency[num]
        mode = num

print("Mode:", mode)
```

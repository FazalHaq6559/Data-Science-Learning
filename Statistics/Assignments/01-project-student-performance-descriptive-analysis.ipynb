<b style= "color:#003366">Store Each Column in Separate Lists


```python
# Data stored in separate lists
gender = ["female", "female", "female", "male", "male", "female", "female", "male", "male", "female",
          "male", "male", "female", "male", "female"]

race = ["group B", "group C", "group B", "group A", "group C", "group B", "group B", "group B", "group D", "group B",
        "group C", "group D", "group B", "group A", "group A"]

parental_education = ["bachelor's degree", "some college", "master's degree", "associate's degree", "some college",
                      "associate's degree", "some college", "some college", "high school", "high school",
                      "associate's degree", "associate's degree", "high school", "some college", "master's degree"]

lunch = ["standard", "standard", "standard", "free", "standard", "standard", "standard", "free", "free", "free",
         "standard", "standard", "standard", "standard", "standard"]

course = ["none", "completed", "none", "none", "none", "none", "completed", "none", "completed", "none",
          "none", "none", "none", "completed", "none"]

math_score = [72, 69, 90, 47, 76, 71, 88, 40, 64, 38, 58, 40, 65, 78, 50]
reading_score = [72, 90, 95, 57, 78, 83, 95, 43, 64, 60, 54, 52, 81, 72, 53]
writing_score = [74, 88, 93, 44, 75, 78, 92, 39, 67, 50, 52, 43, 73, 70, 58]
```

### 1Ô∏è Find the **mean** of `math score`, `reading score`, `writing score`.  


```python
# We use the statistics.mean() function for average values.

import statistics as stats

mean_math = stats.mean(math_score)
mean_reading = stats.mean(reading_score)
mean_writing = stats.mean(writing_score)

print("Mean Math Score:", mean_math)
print("Mean Reading Score:", mean_reading)
print("Mean Writing Score:", mean_writing)
```

### 2Ô∏è‚É£ Find the **median** of `math score`, `reading score`, `writing score`.  
 


```python
# Median is the middle value when data is sorted.
median_math = stats.median(math_score)
median_reading = stats.median(reading_score)
median_writing = stats.median(writing_score)

print("Median Math Score:", median_math)
print("Median Reading Score:", median_reading)
print("Median Writing Score:", median_writing)
```

### 3Ô∏è‚É£ Find the **mode** of all columns (including categorical variables).  


Mode just means the most frequent value, and that concept works for both:

Numerical data ‚Üí e.g., most common math score.

Categorical data ‚Üí e.g., most common gender, lunch type, race, etc.

```python
# Mode is the most frequent value.

mode_math = stats.mode(math_score)
mode_reading = stats.mode(reading_score)
mode_writing = stats.mode(writing_score)

mode_gender = stats.mode(gender)
mode_race = stats.mode(race)
mode_parental_education = stats.mode(parental_education)
mode_lunch = stats.mode(lunch)
mode_course = stats.mode(course)

print("Mode Math Score:", mode_math)
print("Mode Reading Score:", mode_reading)
print("Mode Writing Score:", mode_writing)
print("Mode Gender:", mode_gender)
print("Mode Race:", mode_race)
print("Mode Parental Education:", mode_parental_education)
print("Mode Lunch:", mode_lunch)
print("Mode Course:", mode_course)
```

### 4Ô∏è‚É£ Find the **standard deviation** of `math score`, `reading score`, `writing score`.  


```python
# Standard deviation shows spread of data.

std_math = stats.stdev(math_score)
std_reading = stats.stdev(reading_score)
std_writing = stats.stdev(writing_score)

print("Standard Deviation Math Score:", std_math)
print("Standard Deviation Reading Score:", std_reading)
print("Standard Deviation Writing Score:", std_writing)
```

### 5Ô∏è‚É£ Write a **short note** about **box plot** and **pie chart**. 

**Box Plot:**
A box plot (whisker plot) shows the minimum, first quartile, median, third quartile, and maximum values, making it easy to see data spread and outliers.

**Pie Chart:**
A pie chart shows proportions as slices of a circle, helping visualize percentage distribution across categories.
